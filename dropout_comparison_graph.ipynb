```python
import matplotlib.pyplot as plt
import numpy as np

dataArray0 = np.genfromtxt('sgd_128_0.0001_dropout=0.csv', delimiter=',', names=True)
dataArray2 = np.genfromtxt('sgd_128_0.0001_dopout=0.2.csv', delimiter=',', names=True)
dataArray4 = np.genfromtxt('sgd_128_0.0001_dropout=0.4.csv', delimiter=',', names=True)
dataArray6 = np.genfromtxt('sgd_128_0.0001_dropout=0.6.csv', delimiter=',', names=True)
acc0=dataArray0["test_acc"]
acc2=dataArray2["test_acc"]
acc2=acc2[:120]
acc4=dataArray4["test_acc"]
acc6=dataArray6["test_acc"]
plt.plot(range(120),acc0,'-',linewidth=3,label='dropout=0')
plt.plot(range(120),acc2,'-',linewidth=3,label='dropout=0.2')
plt.plot(range(120),acc4,'-',linewidth=3,label='dropout=0.4')
plt.plot(range(120),acc6,'-',linewidth=3,label='dropout=0.6')
plt.legend()
#plt.ylim(70,92)
plt.xlabel('epoch')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```
